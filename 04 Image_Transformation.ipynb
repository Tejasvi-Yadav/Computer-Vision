### image shif/transformation

```python
import cv2
import numpy as np

image = cv2.imread("lenna.png")

# store height and width of the image
height, width = image.shape[:2]

quarter_height, quarter_width = height / 4, width/ 4

T = np.float32([[2,0,quarter_width],[0,2,quarter_height]])

# WE USE warpAffine to transform the image using the matrix, T
img_translation = cv2.warpAffine(image, T , (width, height))

cv2.imshow("Originalimage", image)
cv2.imshow("Translation", img_translation)
cv2.waitKey()

cv2.destroyAllWindows()
```

### Image Rotation

```python
import cv2 

# load the image
image = cv2.imread("lenna.png")

# Rotate the image 45 degrees clockwise
rotated_image = cv2.rotate(image, cv2.ROTATE_90_COUNTERCLOCKWISE)

# show the rotated image
cv2.imshow("Rotated Image",rotated_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2 

# load the image
image = cv2.imread("lenna.png")

# resize the image to a fized width and height
resize_image = cv2.resize(image, (400,600))

# resize the image with aspect ration preserved
resize_image = cv2.resize(image,None, fx=0.5,fy=0.5)

# show the rotated image
cv2.imshow("Rotated Image",resize_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Using scikit-image

```python
from skimage.transform import resize

# load the image
image = cv2.imread("lenna.png")

# resize with bilinear interpolation
resize_image = resize(image, (1100,1100),anti_aliasing = True)

# resize with bicubic interpolation
resize_image = resize(image, (1200,1200),anti_aliasing = True, order = 3)

# show the resized image
cv2.imshow("Resized Image",resize_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Image Flipping

```python
import cv2 

# load the image
image = cv2.imread("lenna.png")

# Flip the image horizontally
flipped_horizontal = cv2.flip(image,0)

# Flip the image vertically
flipped_vertical = cv2.flip(image,1)

# Flip the image diagonally
flipped_diagonal = cv2.flip(image,-1)

# show the flipped image
cv2.imshow("Original",image)
cv2.imshow("Horizontal Flip",flipped_horizontal)
cv2.imshow("Vertical Flip",flipped_vertical)
cv2.imshow("Diagonal Flip",flipped_diagonal)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### ### Image Flipping using PIL

```python
from PIL import Image

# load the image
image = Image.open("lenna.png")

# Flip the image horizontally
flipped_image = image.transpose(Image.FLIP_LEFT_RIGHT)
flipped_image.show()

# Flip the image vertically
flipped_image = image.transpose(Image.FLIP_TOP_BOTTOM)
flipped_image.show()

# Flip both axes diagonally flip
flipped_image = image.transpose(Image.TRANSPOSE)
flipped_image.show()

```
