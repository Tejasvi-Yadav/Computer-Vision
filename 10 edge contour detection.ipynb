```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
image = cv2.imread("lenna.png",0)
```

```python
image.shape
```

### sobel operator

```python
#sobel parameters:ksize , dx,dy,ddepth
sobelx = cv2.Sobel(image,cv2.CV_64F, 1,0, ksize=3)  #sobel for horizontal edges
sobely = cv2.Sobel(image,cv2.CV_64F,0,1, ksize=3)   #sobel for  vertical edges

#Combine sobel images
combined_sobel = cv2.add(sobelx, sobely)
```

```python
thresh, binary_sobel = cv2.threshold(combined_sobel,100,255,cv2.THRESH_BINARY)
```

```python
plt.figure(figsize=(10,6))
plt.subplot(1,4,1)
plt.title("Original")
plt.imshow(image,cmap='gray')
plt.subplot(1,4,2)
plt.title("Sobel X")
plt.imshow(sobelx,cmap='gray')
plt.subplot(1,4,3)
plt.title("Sobel y")
plt.imshow(sobely,cmap='gray')
plt.subplot(1,4,4)
plt.title("final")
plt.imshow(combined_sobel,cmap='gray')
```

### scharr 

```python
fieldx = cv2.Scharr(image,cv2.CV_32F,1,0)/15.34

fieldy = cv2.Scharr(image,cv2.CV_32F,0,1)/15.34

plt.figure(figsize=(10,6))
plt.subplot(1,3,1)
plt.title("Original")
plt.imshow(image,cmap='gray')
plt.subplot(1,3,2)
plt.title("Y-value")
plt.imshow(fieldy,cmap='gray')
plt.subplot(1,3,3)
plt.title("X-value")
plt.imshow(fieldx,cmap='gray')
```

```python
cv2.imshow('Ximage',fieldx)
cv2.imshow('Yimage',fieldy)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.imshow('Ximage',sobelx)
cv2.imshow('Yimage',sobely)
cv2.imshow('combined',combined_sobel)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### canny edge 

```python
import cv2

img = cv2.imread('rgy.jpg',0)  #load as gray scale

edges = cv2.Canny(img,50,150)

cv2.imshow('Original image',img)
cv2.imshow('Canny edge detection',edges)
cv2.waitKey(0)
cv2.destroyAllWindows()

```

### Contours Detection

```python
import cv2
img = cv2.imread('rgy.jpg')

gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

#creates a binary image where pixels above 127 are set to 255(white) and others to 0(black)
thresh = cv2.threshold(gray,127,255,cv2.THRESH_BINARY)[1]

#cv2.findContours finds contours in binary image
#RETR_TREE retrives all contours and creates a hierarchy tre
#cv2.CHAIN_APPROX_SIMPLE approximates contours to save memory
contours,hierarchy = cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)

#-1 draws all contours
#(0,255,0) is the color(green)
#2 is the thickness of lines
cv2.drawContours(img,contours,-1,(0,255,0),2)

cv2.imshow('Original image',img)
cv2.imshow('Threshold image',thresh)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
